```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_raw = pd.read_csv('2019PlayoffsNBAData.csv')
df_raw.head()
```

```python
df_raw.isnull().sum()
```

```python
df_raw.describe()
```

```python
minute_filter = df_raw['MP'] >= np.percentile(df_raw['MP'], 25)
df_minute_filter = df_raw.loc[minute_filter]
df_minute_filter
```

```python
df_filtered = df_minute_filter.fillna(0)
df_filtered.set_index('Player', inplace=True)
df_filtered.isnull().sum()
```

```python
df_filtered
```

```python
df_filtered['Pos'] = df_filtered['Pos'].astype('category')
df_filtered['Tm'] = df_filtered['Tm'].astype('category')
df_filtered.info()
```

```python
Pos_series = pd.Series(df_filtered['Pos'])
Age_series = pd.Series(df_filtered['Age'])
Tm_series = pd.Series(df_filtered['Tm'])
G_series = pd.Series(df_filtered['G'])
GS_series = pd.Series(df_filtered['GS'])
```

```python
cols = ['Pos', 'Age', 'Tm', 'G', 'GS']
df_filtered_num = df_filtered.drop(columns=cols)
df_filtered_num
```

```python
df_minute_norm = df_filtered_num.divide(df_filtered_num['MP'], axis=0)
df_minute_norm = df_minute_norm.drop(columns='MP')
df_minute_norm
```

```python
from sklearn.preprocessing import StandardScaler
df_scaled = pd.concat([Age_series, df_minute_norm], axis=1)
scaler_values = StandardScaler().fit_transform(df_scaled.values)
scaler_values.shape
```

```python
from sklearn.cluster import KMeans
kmm  = KMeans(n_clusters=5, random_state=1)
kmm.fit(scaler_values)
labels = kmm.labels_
labels
```

```python
from sklearn.decomposition import PCA
pca2 = PCA(2)
decomp = pca2.fit_transform(scaler_values)
```

```python
sns.scatterplot(x=decomp[:,0], y=decomp[:,1], hue=labels, alpha=0.7)
```

```python
df_filtered['Cluster'] = labels
df_filtered.head()
```

```python
df_clustered = df_filtered.copy()
df_clustered.loc['Giannis Antetokounmpo', 'Cluster']
```

```python
cluster2 = df_clustered['Cluster'] == 2
df_cluster2 = df_clustered[cluster2]
df_cluster2
```

```python
df_clustered.columns
```

```python
gb1 = df_clustered.groupby('Cluster')[['PTS', 'FG%', '2P%', '3P%', 'AST']].mean()
gb2 = df_clustered.groupby('Cluster')['Cluster'].count()
cluster_summary = pd.concat([gb1, gb2], axis=1)
cluster_summary
```
